```python
%pylab inline
from scipy.optimize import curve_fit
```

```python
data1 = loadtxt('testrun_25oC_8thrun_8DROPS_15mL_Banging_NOBELPRIZE_ENDGAME_LVL69_BOSS_NOSCOPE_HACKER_NOOB_gooddata.dat')
data2 = loadtxt('testrun_26oC_4_8DROPS_15mL_slightovershoot.dat')
data3 = loadtxt('testrun_26oC_2.dat')
```

```python
time1 = data1[:,0]
conductance1 = data1[:,1]
time2 = data2[:,0]
conductance2 = data2[:,1]
time3 = data3[:,0]
conductance3 = data3[:,1]
```

```python

plot(time1,conductance1)
plot(time2,conductance2,marker='^',linestyle='',color='black')
plot(time3,conductance3,marker='^',linestyle='')
title('The change of conductance against time under 299K')
xlabel('time/s')
ylabel('Conductance/uS')
show()
```

After this 3 repeats, the blue curve seemed to be the best representation for the reaction,while green curve is discarded.

```python
figure(figsize=(10,10))
plot(time1,conductance1, label = 'raw_data')
title('Graph to show progression of reaction, raw data and curve of best fit')

def conductance_f(time1, k, l_inf, l_zero):
    "Blackbody radiation intensity"
    return (l_inf -(l_inf - l_zero)*exp(-k*time1))

curve_est = curve_fit(conductance_f, time1, conductance1, p0 = [0.0005,1700,0])
params = curve_est[0]
k_1 = params[0]
l_inf = params[1]
l_zero = params[2]

time_g = linspace(0,700, 1000)
y_est = l_inf -(l_inf - l_zero)*exp(-k_1*time_g)

plot(time_g, y_est, label = 'curve_fit', color = 'orange')
legend(prop={'size': 25})
xlabel('time/s')
ylabel('Conductance/uS')
show()
```

```python
k_std_1=(curve_est[1][0,0])**0.5
k_std_1,k_1
```

```python

x_temp=1/Temperature
y_rate=log(k_data)
polyfit(x_temp,y_rate,1,cov=True)
param=polyfit(x_temp,y_rate,1,cov=True,w=1/k_err**2)
m=param[0][0]
c=param[0][1]
predicted_y=m*x_temp+c
plot(x_temp,y_rate,linestyle='',marker='o')
plot(x_temp,predicted_y,linestyle='-',marker='')
error=s[:,2]/k_data
errorbar(x_temp,y_rate,linestyle='',yerr=error,markersize='2.5')
title('Rate against temperature',fontsize=18)
xlabel('1/T(K^(-1))')
ylabel('ln(k)per second')
show()
param
```
